```python
df = pd.read_csv('shenzhen.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df[df['communityName']=='communityName']
```

```python
df1 = df[df['communityName']!='communityName']   # 通过筛选删除特定行
```

```python
df2 = df1.rename(columns={'communityName':'社区名称','areaName':'所在区','roomType':'房屋户型','louCeng':'所在楼层','area':'建筑面积','jianZhu':'建筑类型','chaoXiang':'房屋朝向','zhuangXiu':'装修情况','tiHu':'梯户比例','guapaitime':'挂牌时间','jiaoyitime':'上次交易','totalPrice':'总价','location':'具体地点','nianxian':'房本年限'})
```

```python
df2.info()
```

```python
df2.isnull().any()  # 查看每列是否有缺失值
```

```python
df2.isnull().sum()  # 查看每列是否有缺失值
```

```python
df3 = df2.dropna()  # 删除所有含有缺失值的行
```

```python
df3
```

```python
df3['建筑面积'] = df3['建筑面积'].str.extract('(\d+.?\d+)㎡', expand=False)    # 正则表达式提取信息
```

```python
df3.head()
```

```python
df3['总价'] = df3['总价'].str.extract('(\d+)万元', expand=False)     # 正则表达式提取信息
```

```python
df3.head()
```

```python
df3[['总价']] = df3[['总价']].astype(int)  # 转换数据类型
df3[['建筑面积']] = df3[['建筑面积']].astype(float)   # 转换数据类型
```

```python
df3['均价'] = df3['总价']/df3['建筑面积']
```

```python
df3.info()
```

```python
df3.isnull().any()  # 查看每列是否有缺失值
```

```python
df3.describe() 
```

```python
df3.to_csv('df3_shenzhen.csv')    # df3 导出
```

```python
f, [ax1,ax2,ax3] = plt.subplots(1,3,figsize=(18,5))
sns.countplot(df3['装修情况'],ax=ax1)
ax1.set_title('深圳二手房装修情况',fontsize=18)
ax1.set_xlabel('装修情况', fontsize=15)
ax1.set_ylabel('数量',fontsize=15)

sns.barplot(x='装修情况', y='总价', data=df3,ax=ax2)
ax2.set_title('装修情况与房屋总价分布',fontsize=18)
ax2.set_xlabel('装修情况', fontsize=15)
ax2.set_ylabel('房屋总价', fontsize=15)

sns.boxplot(x='装修情况', y='总价', data=df3,ax=ax3)
ax3.set_title('装修情况与房屋总价分布',fontsize=18)
ax3.set_xlabel('装修情况', fontsize=15)
ax3.set_ylabel('房屋总价', fontsize=15)

plt.savefig('img1')
```

```python
f, ax1 = plt.subplots(figsize=(8,5))

sns.regplot(x='建筑面积', y='总价', data=df3,ax=ax1)
ax1.set_title('房屋面积与总价关系',fontsize=18)
ax1.set_xlabel('面积(㎡)', fontsize=15)
ax1.set_ylabel('总价(万元)', fontsize=15)
plt.savefig('img2')
```

```python
f, ax1 = plt.subplots(figsize=(8,5))

data = df3['建筑面积']
sns.distplot(data,ax=ax1,color='r')
sns.kdeplot(data,ax=ax1,shade=True)  # 拟合并绘制单变量核密度估计图
ax1.set_title('建筑面积核密度估计图',fontsize=18)
ax1.set_xlabel('面积(㎡)',fontsize=15)

plt.savefig('img3')
```

```python
f, ax1 = plt.subplots(figsize=(12,12))
sns.countplot(y='房屋户型',data=df3,ax=ax1)
ax1.set_title('房屋户型',fontsize=18)
ax1.set_xlabel('数量',fontsize=15)

plt.savefig('img4')
```

```python
df_sz = pd.read_csv('df3_shenzhen.csv')
```

```python
df_house_count = df_sz.groupby('所在区')['Unnamed: 0'].count().sort_values(ascending=False).to_frame().reset_index().rename(columns={'Unnamed: 0':'数量'})
```

```python
df_house_count
```

```python
df_house_mean = df3.groupby('所在区')['均价'].mean().sort_values(ascending=False).to_frame().reset_index()
```

```python
df_house_mean
```

```python
f, [ax1,ax2,ax3] = plt.subplots(3,1,figsize=(10,20))

sns.barplot(x='所在区',y='数量',data=df_house_count,ax=ax1)
ax1.set_title('深圳各区二手房数量对比',fontsize=18)
ax1.set_xlabel('区域',fontsize=15)
ax1.set_ylabel('数量',fontsize=15)

sns.barplot(x='所在区',y='均价',data=df_house_mean,ax=ax2)
ax2.set_title('深圳各区二手房每平米单价对比',fontsize=18)
ax2.set_xlabel('区域',fontsize=15)
ax2.set_ylabel('每平米单价（万元）',fontsize=15)

sns.boxplot(x='所在区',y='总价',data=df_sz,ax=ax3)
ax3.set_title('深圳各区二手房房屋总价',fontsize=18)
ax3.set_xlabel('区域',fontsize=15)
ax3.set_ylabel('房屋总价（万元）',fontsize=15)
plt.savefig('img5')
```

```python
f, ax1 = plt.subplots(figsize=(15,26))
sns.countplot(y='梯户比例',data=df_sz,ax=ax1)
ax1.set_title('深圳二手房梯户比例',fontsize=18)
ax1.set_xlabel('数量', fontsize=15)
ax1.set_ylabel('梯户比例',fontsize=15)

plt.savefig('img6')
```
