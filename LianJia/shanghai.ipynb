```python
df2['建筑面积'] = df2['建筑面积'].str.extract('(\d+.\d+)㎡', expand=False)    # 正则表达式提取信息
```

```python
df2['总价'] = df2['总价'].str.extract('(\d+)万元', expand=False)     # 正则表达式提取信息
```

```python
df2.head()
```

```python
df2[['总价']] = df2[['总价']].astype(int)  # 转换数据类型
```

```python
df2[['建筑面积']] = df2[['建筑面积']].astype(float)   # 转换数据类型
```

```python
df2.info()
```

```python
df2['均价'] = df2['总价']/df2['建筑面积']
```
